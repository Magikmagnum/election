# Normalisation

1. Renommer correctement

```python
# Renommer "Libellé" en "code_departement"
df = df.rename(columns={"Libellé": "code_dept"})
```

2. Transformer (O) en NaN

```python
df.replace("(O)", np.nan, inplace=True)
df.info()
```

3. Aplatir le DataFrame (melt)

```python
df_long = df.melt(
    id_vars="code_dept",
    var_name="periode",
    value_name="taux_chomage"
)
```

5. Extraire année et trimestre

```python
df_long["annee"] = df_long["periode"].str.slice(0, 4).astype(int)
df_long["trimestre"] = df_long["periode"].str[-1].astype(int)

df_long.drop(columns="periode", inplace=True)
df_long.head()
```

```python
df_long.shape
```

6. Insertion SQLAlchemy

```python
from sqlalchemy.orm import Session
from models import Chomage
from db import engine

session = Session(bind=engine)

for row in df_long.itertuples(index=False):
    session.add(
        Chomage(
            code_dept=row.code_dept,
            annee=row.annee,
            trimestre=row.trimestre,
            taux_chomage=row.taux_chomage
        )
    )

session.commit()
session.close()
```
