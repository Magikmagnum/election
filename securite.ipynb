```python
colonnes_essentielles = [
    "Code_departement",
    "annee",
    "indicateur",
    "unite_de_compte",
    "nombre",
    "taux_pour_mille",
]

df_essentiel = df[colonnes_essentielles].copy()
print(df_essentiel.head(1000))
```

```python
df_indicateur = (
    df_essentiel[['indicateur']]
    .drop_duplicates()
    .rename(columns={'indicateur': 'libelle'})
    .sort_values('libelle')
    .reset_index(drop=True)
)

df_indicateur.head(100)
```

```python
df_unite_de_compte = (
    df_essentiel[['unite_de_compte']]
    .drop_duplicates()
    .rename(columns={'unite_de_compte': 'libelle'})
    .sort_values('libelle')
    .reset_index(drop=True)
)

df_unite_de_compte.head(100)
```

```python
from db import SessionLocal
session = SessionLocal()
```

```python
from models import Indicateur

existing = {
    i.libelle
    for i in session.query(Indicateur.libelle).all()
}

indicateurs = [
    Indicateur(libelle=row.libelle)
    for row in df_indicateur.itertuples(index=False)
    if row.libelle not in existing
]

session.bulk_save_objects(indicateurs)
session.commit()
```

```python
from models import UniteDeCompte

existing_unite = {
    u.libelle 
    for u in session.query(UniteDeCompte.libelle).all()
}

unites = [
    UniteDeCompte(libelle=row.libelle)
    for row in df_unite_de_compte.itertuples(index=False)
    if row.libelle not in existing_unite
]

session.bulk_save_objects(unites)
session.commit()
```

```python
from models import Departement

departements = {
    d.code_departement: d.id
    for d in session.query(Departement).all()
}
```

```python
def adapt_departement(code_df):
    """
    Transforme un code département du DataFrame
    en code compatible avec la base.

    Exemple :
    - 995  -> ZA
    - 986  -> ZB
    - 1..95 -> '1', '2', ...
    """
    mapping_special = {
        995: "ZA",
        986: "ZB",
        987: "ZC",
        988: "ZD",
        990: "ZE",
        991: "ZF",
        992: "ZG",
        993: "ZH",
        994: "ZI"
    }

    # si le code est dans le mapping spécial
    if code_df in mapping_special:
        return mapping_special[code_df]

    # sinon on garde le code numérique classique
    return str(code_df)
```

```python
from models import FaitSecurite

faits = []

for row in df_essentiel.itertuples(index=False):

    # ===== Adapter le code département =====
    adapted_code = adapt_departement(row.Code_departement)

    departement_id = departements.get(adapted_code)
    indicateur_id = indicateurs.get(row.indicateur)
    unite_id = unites.get(row.unite_de_compte)

    # ===== Vérification des FK : on ignore les lignes incohérentes =====
    if not departement_id or not indicateur_id or not unite_id:
        continue

    # ===== Création du FaitSecurite =====
    faits.append(
        FaitSecurite(
            departement_id=departement_id,
            annee=row.annee,
            indicateur_id=indicateur_id,
            unite_de_compte_id=unite_id,
            nombre=row.nombre,
            taux_pour_mille=row.taux_pour_mille
        )
    )

# ===== Insertion en base =====
session.bulk_save_objects(faits)
session.commit()
```
