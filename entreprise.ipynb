```python
import numpy as np
import pandas as pd
import os
os.listdir()
```

```python
df_echanges = pd.read_excel('./data/economie/entreprise/echanges-extérieurs-ecc-base-2020.xlsx')
df_creation_entreprise = pd.read_excel('./data/economie/entreprise/creations-entreprises.xlsx')
```

```python
df_echanges.head()
```

```python
df_echanges.info()
```

```python
columns = ["annee", "exportations", "importations"]
df_echanges.columns = columns
df_echanges.head()
```

```python
df_creation_entreprise.head()
```

```python
columns_entreprise = ["trimestre", "entreprise", "entrepreneurs"]
df_creation_entreprise.columns = columns_entreprise
df_creation_entreprise.head(50)
```

```python
df_creation_entreprise['annee'] = df_creation_entreprise['trimestre'].str[:4]
df_entreprise = df_creation_entreprise.groupby('annee')[['entreprise','entrepreneurs']].sum().reset_index()
df_entreprise[['entreprise','entrepreneurs']] = df_entreprise[['entreprise','entrepreneurs']]*1000
df_entreprise.head(50)
```

```python
df_entreprise['micro_entrepreneurs'] = df_entreprise['entreprise'] - df_entreprise['entrepreneurs']
```

```python
df_entreprise.head(50)
```

```python
df_entreprise_essenciel = df_entreprise.drop(columns=['entreprise'])
df_entreprise_essenciel.head(50)
```

```python
df_entreprise_essenciel['annee'] = df_entreprise_essenciel['annee'].astype(int)
df_echanges['annee'] = df_echanges['annee'].astype(int)
```

```python
df_final = pd.merge(
    df_entreprise_essenciel,
    df_echanges,
    on='annee',
    how='inner'
)

df_final.head(50)
```

```python
# Ratio micro / total
df_final['total_entreprises'] = df_final['entrepreneurs'] + df_final['micro_entrepreneurs']
df_final['ratio_micro'] = df_final['micro_entrepreneurs'] / df_final['total_entreprises']

# Ratio export / import (solde commercial relatif)
df_final['solde_commercial'] = df_final['exportations'] - df_final['importations']
df_final['ratio_export_import'] = df_final['exportations'] / df_final['importations']

# Croissance annuelle des variables clés
df_final['croissance_total_entreprises'] = df_final['total_entreprises'].pct_change()  # % de variation
df_final['croissance_micro'] = df_final['micro_entrepreneurs'].pct_change()
df_final['croissance_export'] = df_final['exportations'].pct_change()
df_final['croissance_import'] = df_final['importations'].pct_change()
```

```python
df_final.head(50)
```

```python
df = pd.DataFrame()
df = df_final[(df_final["annee"] >= 2007) & (df_final["annee"] <= 2022)]
df.head(50)
```

```python
df_model = df[['annee',
               'ratio_micro',          # proportion des micro-entrepreneurs
               'solde_commercial',     # export - import
               'ratio_export_import',  # export/import
               'croissance_total_entreprises',  # croissance des entreprises hors micro
               'croissance_micro',     # croissance des micro-entrepreneurs
               'croissance_export',    # croissance exportations
               'croissance_import'     # croissance importations
              ]].copy()

# Remplacer NaN (venant de pct_change) par 0
df_model.fillna(0, inplace=True)
df_model.head(50)
```

```python
df_entreprise_contexte = df_model[[ 'annee', 
                            'croissance_total_entreprises',
                            'ratio_micro',
                            'solde_commercial',
                            'croissance_export'
                        ]].copy()

df_entreprise_contexte.head(50)
```

```python
from db import SessionLocal
from models import Entreprise

# =====================================================
# CRÉATION DE LA SESSION
# =====================================================
session = SessionLocal()

# =====================================================
# INSERTION DES DONNÉES
# =====================================================
for row in df_entreprise_contexte.itertuples(index=False):
    record = Entreprise(
        annee=int(row.annee),
        croissance_total_entreprises=float(row.croissance_total_entreprises)
            if row.croissance_total_entreprises is not None else None,
        ratio_micro=float(row.ratio_micro)
            if row.ratio_micro is not None else None,
        solde_commercial=float(row.solde_commercial)
            if row.solde_commercial is not None else None,
        croissance_export=float(row.croissance_export)
            if row.croissance_export is not None else None,
    )

    session.add(record)

# =====================================================
# COMMIT
# =====================================================
session.commit()
session.close()

print("Insertion terminée : entreprise_contexte")
```
