```python
def split_and_explode_population(df):
    """
    S√©pare le dataframe source en hommes / femmes,
    puis transforme chaque sous-dataframe en format long
    """

    # --- 1. S√©paration hommes / femmes ---
    df_homme = pd.concat([df.iloc[:, :2], df.iloc[:, 8:14]], axis=1)
    df_femme = pd.concat([df.iloc[:, :2], df.iloc[:, 14:20]], axis=1)

    return df_homme, df_femme
```

```python
import pandas as pd
import math

def explode_population_dataframe(df, sexe, annee, debug=True):
    """
    Transforme un dataframe large (par tranche d'√¢ge)
    en dataframe long (1 ligne = 1 tranche)
    """
    lignes = []

    tranche_map = {
        2: "0 √† 19 ans",
        3: "20 √† 39 ans",
        4: "40 √† 59 ans",
        5: "60 √† 74 ans",
        6: "75 ans et plus"
    }

    if debug:
        print("=== D√âBUT TRAITEMENT DATAFRAME ===")
        print(f"Sexe trait√© : {sexe}")
        print(f"Nombre de lignes source : {len(df)}")
        print("-" * 50)

    for idx, row in df.iterrows():

        code_raw = row.iloc[0]
        nom_dept = row.iloc[1]

        # üö´ SKIP ligne d'en-t√™te ou invalide
        if pd.isna(code_raw) or not str(code_raw).isdigit():
            if debug:
                print(f"[LIGNE {idx}] ‚è≠Ô∏è ignor√©e (ligne d'en-t√™te ou invalide)")
            continue

        code_dept = str(code_raw).zfill(2)

        if debug:
            print(f"[LIGNE {idx}] D√©partement {code_dept} - {nom_dept}")

        for col_idx, tranche in tranche_map.items():
            population = row.iloc[col_idx]

            # s√©curit√© suppl√©mentaire
            if pd.isna(population):
                population = 0

            if debug:
                print(
                    f"  ‚Üí Tranche: {tranche:<15} | "
                    f"Population: {population}"
                )

            lignes.append({
                "code_dept": code_dept,
                "nom_dept": nom_dept,
                "sexe": sexe,
                "tranche_age": tranche,
                "population": int(population),
                "annee": annee
            })

        if debug:
            print("-" * 50)

    df_result = pd.DataFrame(lignes)

    if debug:
        print("=== FIN TRAITEMENT ===")
        print(f"Nombre total de lignes g√©n√©r√©es : {len(df_result)}")
        print("Aper√ßu du r√©sultat :")
        print(df_result.head())

    return df_result
```

```python
import pandas as pd

df_population_all = []

for annee in range(2007, 2023):
    print(f"\n=== Traitement population {annee} ===")

    # 1. Lecture du fichier Excel
    df = pd.read_excel(f"./populations/pop-{annee}.xlsx")

    # Affichage rapide des premi√®res lignes
    # print("Aper√ßu du fichier brut :")
    # print(df.head(2))
    
    # 2. S√©paration hommes / femmes
    df_homme, df_femme = split_and_explode_population(df)

    # 3. Explosion (format long) avec debug activ√©
    df_homme_long = explode_population_dataframe(
        df_homme,
        sexe="H",
        annee=annee,
        debug=False
    )

    df_femme_long = explode_population_dataframe(
        df_femme,
        sexe="F",
        annee=annee,
        debug=False
    )

    # 4. Concat√©ner hommes + femmes pour cette ann√©e
    df_population_all.append(pd.concat([df_homme_long, df_femme_long], ignore_index=True))

# 5. Concat√©nation finale sur toutes les ann√©es
df_population = pd.concat(df_population_all, ignore_index=True)


print("\n=== Aper√ßu final population ===")
print(df_population.head(15))
print("\nNombre total de lignes :", len(df_population))
```

```python
def seed_sexe(session):
    codes = ["ENSEMBLE", "H", "F"]
    for code in codes:
        if not session.query(Sexe).filter_by(code=code).first():
            session.add(Sexe(code=code))
```

```python
def seed_tranche_age(session):
    tranches = [
        (0, 19),
        (20, 39),
        (40, 59),
        (60, 74), 
        (75, None),
    ]

    for age_min, age_max in tranches:
        exists = session.query(TrancheAge).filter_by(
            age_min=age_min,
            age_max=age_max
        ).first()
        if not exists:
            session.add(
                TrancheAge(age_min=age_min, age_max=age_max)
            )
```

```python
def run():
    session = SessionLocal()
    try:
        seed_sexe(session)
        seed_tranche_age(session)
        session.commit()
    except:
        session.rollback()
        raise
    finally:
        session.close()
```

```python
from sqlalchemy.orm import Session
import re
from models import PopulationDepartement, Departement
```

```python
# Mapping direct tranche -> (age_min, age_max)
TRANCHES_AGE_MAP = {
    "0 √† 19 ans": (0, 19),
    "20 √† 39 ans": (20, 39),
    "40 √† 59 ans": (40, 59),
    "60 √† 74 ans": (60, 74),
    "75 ans et plus": (75, None),
}

def parse_tranche_age(tranche_str: str):
    """
    Retourne (age_min, age_max) √† partir d'une cha√Æne de tranche connue.
    """
    return TRANCHES_AGE_MAP[tranche_str]
```

```python
def inserer_population(df, session: Session):
    # Mapping direct des tranches d'√¢ge
    TRANCHES_AGE_MAP = {
        "0 √† 19 ans": (0, 19),
        "20 √† 39 ans": (20, 39),
        "40 √† 59 ans": (40, 59),
        "60 √† 74 ans": (60, 74),
        "75 ans et plus": (75, None),
    }

    # Mapping des d√©partements d'outre-mer
    DOM_CODES_MAP = {
        "971": "ZA",  # Guadeloupe
        "972": "ZB",  # Martinique
        "973": "ZC",  # Guyane
        "974": "ZD",  # La R√©union
    }

    # R√©cup√©ration des objets existants
    departements = {d.code_dept: d for d in session.query(Departement).all()}
    sexes = {s.code: s for s in session.query(Sexe).all()}
    tranches = session.query(TrancheAge).all()
    tranches_map = {(t.age_min, t.age_max): t for t in tranches}

    # Pr√©charger les populations existantes pour √©viter doublons
    existing_keys = set(
        session.query(
            PopulationDepartement.departement_code,
            PopulationDepartement.sexe_id,
            PopulationDepartement.tranche_age_id,
            PopulationDepartement.annee
        ).all()
    )

    to_insert = []

    for _, row in df.iterrows():
        # Normalisation du code d√©partement
        code_dept_df = str(row["code_dept"]).zfill(3)  # ex: 971 ‚Üí "971"
        code_dept_db = DOM_CODES_MAP.get(code_dept_df, code_dept_df)

        dept = departements.get(code_dept_db)
        sexe = sexes.get(str(row["sexe"]).strip())

        if not dept:
            print(f"‚ö†Ô∏è D√©partement non trouv√© pour la ligne: {row['code_dept']}")
            continue

        if not sexe:
            print(f"‚ö†Ô∏è Sexe non trouv√© pour la ligne: {row['sexe']}")
            continue

        # R√©cup√©rer la tranche d'√¢ge
        age_min, age_max = TRANCHES_AGE_MAP[row["tranche_age"]]
        tranche = tranches_map.get((age_min, age_max))
        if not tranche:
            print(f"‚ö†Ô∏è Tranche d'√¢ge non trouv√©e pour: {row['tranche_age']}")
            continue

        # Cl√© unique pour v√©rifier doublon
        key = (dept.code_dept, sexe.id, tranche.id, row["annee"])
        if key in existing_keys:
            continue  # Ignorer si d√©j√† pr√©sent

        # Cr√©ation de l'objet PopulationDepartement
        pop_dept = PopulationDepartement(
            departement_code=dept.code_dept,
            departement=dept,
            sexe_id=sexe.id,
            sexe=sexe,
            tranche_age_id=tranche.id,
            tranche_age=tranche,
            annee=row["annee"],
            population=row["population"]
        )
        to_insert.append(pop_dept)
        existing_keys.add(key)  # √âviter doublon dans le batch

    # Insert en bloc
    if to_insert:
        session.bulk_save_objects(to_insert)
        session.commit()
        print(f"‚úÖ {len(to_insert)} populations ins√©r√©es avec succ√®s.")
    else:
        print("‚ÑπÔ∏è Aucune nouvelle population √† ins√©rer, tout est d√©j√† pr√©sent.")
```
