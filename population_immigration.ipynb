Renommer les colonnes 

```python
df.columns = [
    "annee", "pct_immigration"
]
```

Normalisation structurelle

```python
df["annee"] = (
    df["annee"]
    .astype(str)
    .str.extract(r"(\d{4})")
    .astype(int)
)
```

```python
df = df[(df["annee"] >= 2007) & (df["annee"] <= 2022)]
```
